```python
import matplotlib.pyplot as plt
# Assign values to phis
phi0 = 1
phi1 = 2
phi2 = 3
# Assuming phi0, phi1, and phi2 are defined as before
x_1 = range(-10, 11)
x_2 = range(-10, 11)
X_1, X_2 = np.meshgrid(x_1, x_2)
Y = phi0 + phi1 * X_1 + phi2 * X_2

plt.contour(X_1, X_2, Y)
plt.xlabel('x_1')
plt.ylabel('x_2')
plt.title('Contour Plot of y = phi0 + phi1*x_1 + phi2*x_2')
plt.colorbar(label='y')
plt.show()
```

```python
import plotly.graph_objects as go


Y = phi0 + phi1 * X_1 + phi2 * X_2

fig = go.Figure(data=[go.Surface(z=Y, x=X_1, y=X_2)])
fig.update_layout(title='Interactive 3D Plot of y = phi0 + phi1*x_1 + phi2*x_2',
                  autosize=False,
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))
fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np

# Assign values to phis
phi0 = 1
phi1 = 2
phi2 = 3

# Assuming phi0, phi1, and phi2 are defined as before
x_1 = np.arange(-10, 11)  # Convert range to NumPy array
x_2 = np.arange(-10, 11)  # Convert range to NumPy array
X_1, X_2 = np.meshgrid(x_1, x_2)
Y = phi0 + phi1 * X_1 + phi2 * X_2

fig = go.Figure(data =
    go.Contour(
        z=Y,
        x=x_1,
        y=x_2
    ))
fig.update_layout(title='Interactive Contour Plot of y = phi0 + phi1*x_1 + phi2*x_2')
fig.show()
```
